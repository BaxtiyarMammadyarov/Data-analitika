```python
v_data={'city':city_gr.index,'count':city_gr.values}
v_df=pd.DataFrame(data=v_data)
v_df
v_list=[]
mx=v_df['count'].max()
mn=v_df['count'].min()
for i in v_df.index:
    if(v_df['count'].iloc[i]==mx):
       v_list.append('max')
    elif (v_df['count'].iloc[i]==mn):
        v_list.append('min')
    else:
        v_list.append('--')
       
v_df['catagory']=pd.DataFrame(v_list)
v_df
```
