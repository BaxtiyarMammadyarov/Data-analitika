```python
plt.plot(
    fb[fb.index.year==2021].index.values,
    fb[fb.index.year==2021].High,
    color="blue",
    linewidth=0.3,
    linestyle="--",
    marker="o",
    alpha=1)

plt.plot(
    fb[fb.index.year==2021].index.values,
    fb[fb.index.year==2021].Low,
    color="red",
    linewidth=0.3,
    linestyle="--",
    marker="o",
    alpha=1)
plt.xticks(rotation=50)
plt.figure(figsize=(50,10))
plt.show()
```

```python
fb_2021=fb[fb.index.year==2021]
fb_2021
```

```python


df=fb_2021.groupby(by=fb_2021.index.month).mean()


Apr=df[df.index==4]['High'].mean()/df[df.index==4]['Open'].mean()
May=df[df.index==5]['High'].mean()/df[df.index==4]['Open'].mean()
Jun=df[df.index==6]['High'].mean()/df[df.index==4]['Open'].mean()
Jul=df[df.index==7]['High'].mean()/df[df.index==4]['Open'].mean()
Aug=df[df.index==8]['High'].mean()/df[df.index==4]['Open'].mean()
Sep=df[df.index==9]['High'].mean()/df[df.index==4]['Open'].mean()
Okt=df[df.index==10]['High'].mean()/df[df.index==4]['Open'].mean()
Nov=df[df.index==11]['High'].mean()/df[df.index==4]['Open'].mean()
Dec=df[df.index==12]['High'].mean()/df[df.index==4]['Open'].mean()
df.index.values
```

```python
plt.scatter(df.index.values,[Apr,May,Jun,Jul,Aug,Sep,Okt,Nov,Dec],c='r')

plt.show()
```

```python
plt.pie(df.index.values
, labels=[
                f'Apr\n({round(Apr, 2)})',
                f'May\n({round(May, 2)})',
                f'Jun\n({round(Jun, 2)})',
                f'Jul\n({round(Jul, 2)})',
                f'Aug\n({round(Aug, 2)})',
                f'Sep\n({round(Sep, 2)})',
                f'Okt\n({round(Okt, 2)})',
                f'Nov\n({round(Nov, 2)})',
                f'Dec\n({round(Dec, 2)})'] ,radius=2)

plt.figure(figsize=(16,9))

plt.show()
```
