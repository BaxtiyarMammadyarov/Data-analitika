```python
filt_lat=ws_data['lat']<0
lat_abs=ws_data[filt_lat]['lat'].apply(lambda s : abs(s))
ws_data.update(lat_abs)
ws_data[filt_lat]
```

```python
filt_lng=ws_data['lng']<0

new_ws_lng=ws_data[filt_lng]['lng'].apply(lambda s : abs(s))
new_ws_lng

ws_data.update(new_ws_lng)
ws_data
```

```python
population_avg=ws_data['population'].mean()
population_avg
```

```python


new_p_data=ws_data[ws_data['population'].isna()]['population'].apply(lambda s : population_avg)
new_p_data

ws_data.update(new_p_data)
ws_data['population'].isna().sum()
```

```python
primary_filt_ws_data=ws_data['capital']=='primary'

ws_data_sort_p=ws_data[primary_filt_ws_data].sort_values(['population'])
ws_data_sort_p
```
